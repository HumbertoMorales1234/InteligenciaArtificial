```python
import cv2 as cv

rostro = cv.CascadeClassifier('haarcascade_frontalface_alt.xml')
cap = cv.VideoCapture(0)
i=0
while True:
    ret, frame = cap.read()
    gray= cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    rostros = rostro.detectMultiScale(gray, 1.3, 5)
    i+=1
    for(x, y, w, h) in rostros:
        frame = cv.rectangle(frame, (x,y), (x+w, y+h), (0, 255, 0), 2)                                                  #Rostro
        frame = cv.rectangle(frame, (x+int(w*3/8),y+int(h*10/24)), (x+int(w*5/8), y+int(h*17/24)), (255, 255, 0), 2)    #Nariz
        frame = cv.circle(frame, ( x+int(w/3.2) , y+int(h/2.3)), 15, (255, 20, 255), 2)                                 #OjoIzq
        frame = cv.circle(frame, ( x+(int(w/3.0)*2) , y+int(h/2.3)), 15, (255, 20, 255), 2)                             #OjoDer
        frame = cv.circle(frame, ( x+(int(w/2)) , y+int(h/1.2)), 20, (40, 40, 255), 2)                                  #Boca
        
        frame2 = frame[x:x+w, y: y+h]
    #cv.imwrite(''+str(i)+'.jpg', frame) #*Guarda los frames del video como imagenes
    cv.imshow('rostros', frame)
    k = cv.waitKey(1)
    if k== ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```

```python
import cv2 as cv

cap = cv.VideoCapture(0)
i=524
while True:
    ret, frame = cap.read()
    frame =  cv.rectangle(frame, (100, 100), (600, 600), (100, 200, 100), 2)
    cara = frame[200:500, 200:500]
    
    frame3= cv.resize(cara, (30,30))
    frame4= cv.rotate(frame3, cv.ROTATE_90_CLOCKWISE)
    frame5= cv.rotate(frame3, cv.ROTATE_90_COUNTERCLOCKWISE)
    frame6= cv.flip(frame3, 0)
    frame7= cv.flip(frame3, 1)
    frame8= cv.flip(frame3, -1)
    
    cv.imshow('cara', cara)
    cv.imshow('resize', frame3)
    cv.imshow('rotado1', frame4)
    cv.imshow('rotado2', frame5)
    cv.imshow('flipped1', frame6)
    cv.imshow('flipped2', frame7)
    cv.imshow('flipped3', frame8)
    
    k = cv.waitKey(1)
    if k == ord('s'):
        i=i+1
        cv.imwrite('imgGeneradas/cara'+str(i)+'.jpg', frame3) #*Guarda los frames del video como imagenes
        i+=1
        cv.imwrite('imgGeneradas/cara'+str(i)+'.jpg', frame4) 
        i+=1
        cv.imwrite('imgGeneradas/cara'+str(i)+'.jpg', frame5) 
        i+=1
        cv.imwrite('imgGeneradas/cara'+str(i)+'.jpg', frame6) 
        i+=1
        cv.imwrite('imgGeneradas/cara'+str(i)+'.jpg', frame7) 
        i+=1
        cv.imwrite('imgGeneradas/cara'+str(i)+'.jpg', frame8) 
        
    if k== ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```
