## Solución Iterativa

Se representan los colores por letras, es decir, todas las casillas con la letra "A" son de un mismo color, todas las "B" son otro y así sucesivamente

```python
class Contador:
    def __init__(self,  color):
        self.color = color
        self.counter = 0
    def increment(self):
        self.counter +=1

class Casilla:
    def __init__(self, color):
        self.color = color
        self.visitado = False
        
def imprime(matriz):
    for x in range(len(matriz)):
        for y in range(len(matriz[0])):
            objeto= matriz[x][y]
            print(objeto.color,'; ', end=' ')
        print("\n")

def mostrarCuentas(contadores):
    for x in range(len(contadores)):
        print("Contador ", contadores[x].color, ': ', contadores[x].counter)

def encontrarContador(contadores, color):
    for x in range(len(contadores)):
        if(contadores[x].color == color):
            contadores[x].increment()
    return contadores

def contar_islas(matriz, contadores):
    filas = len(matriz)
    columnas = len(matriz[0])
    for i in range(filas):
        for j in range(columnas):
            actual = matriz[i][j]
            if(i-1>=0):
                if(matriz[i-1][j].color == actual.color):
                    matriz[i][j].visitado = True
            if(j-1>=0):
                if(matriz[i][j-1].color == actual.color):
                    if(matriz[i][j-1].visitado == True):
                        matriz[i][j].visitado = True
            
            actual= matriz[i][j]
            if(actual.visitado != True):
                if(j+1<columnas):
                    if(matriz[i][j+1].color!= actual.color):
                        contadores= encontrarContador(contadores, actual.color)
                else:
                    contadores= encontrarContador(contadores, actual.color)

    mostrarCuentas(contadores)

contadores = [
    Contador('A'),
    Contador('B'),
    Contador('C'),
    Contador('D'),
    Contador('E'),
]

matriz =[
    [Casilla('A'),Casilla('A'),Casilla('C'),Casilla('C'),],
    [Casilla('A'),Casilla('B'),Casilla('B'),Casilla('B'),],
    [Casilla('A'),Casilla('A'),Casilla('B'),Casilla('E'),],
    [Casilla('D'),Casilla('D'),Casilla('E'),Casilla('E'),],
    [Casilla('A'),Casilla('D'),Casilla('E'),Casilla('E'),],
]

imprime(matriz)
contar_islas(matriz, contadores)
```

## Solución Recursiva

```python
class Contador:
    def __init__(self, color):
        self.color = color
        self.counter = 0

    def increment(self):
        self.counter += 1

class Casilla:
    def __init__(self, color):
        self.color = color
        self.visitado = False

def imprime(matriz):
    for fila in matriz:
        for casilla in fila:
            print(casilla.color, '; ', end=' ')
        print("\n")

def mostrarCuentas(contadores):
    for contador in contadores:
        print("Contador ", contador.color, ': ', contador.counter)


def contar_islas(matriz, contadores, i, j):
    if j >= len(matriz[0]):
        contadores = contar_islas(matriz, contadores, i+1, j=0)
        return contadores
        
    if i >= len(matriz):
        return contadores

    actual = matriz[i][j]
    
    if i - 1 >= 0 and matriz[i - 1][j].color == actual.color:
        matriz[i][j].visitado = True
        contadores = contar_islas(matriz, contadores, i, j+1)
    elif j - 1 >= 0 and matriz[i][j - 1].color == actual.color and matriz[i][j-1].visitado == True:
        matriz[i][j].visitado = True
        contadores = contar_islas(matriz, contadores, i, j + 1)
    elif j + 1 < len(matriz[0]) and matriz[i][j + 1].color != actual.color:
        contadores = encontrarContador(contadores, actual.color)
        contadores = contar_islas(matriz, contadores, i, j + 1)
    elif j + 1 >= len(matriz[0]):
        contadores = encontrarContador(contadores, actual.color)
        contadores = contar_islas(matriz, contadores, i, j + 1)
    else:
        contadores = contar_islas(matriz, contadores, i, j + 1)
    
    return contadores


def encontrarContador(contadores, color):
    for contador in contadores:
        if contador.color == color:
            contador.increment()
    return contadores

contadores = [
    Contador('A'),
    Contador('B'),
    Contador('C'),
    Contador('D'),
    Contador('E'),
]

matriz = [
    [Casilla('A'), Casilla('A'), Casilla('C'), Casilla('C')],
    [Casilla('A'), Casilla('B'), Casilla('B'), Casilla('B')],
    [Casilla('A'), Casilla('A'), Casilla('B'), Casilla('E')],
    [Casilla('D'), Casilla('B'), Casilla('E'), Casilla('E')],
]

imprime(matriz)
contadores = contar_islas(matriz, contadores, 0, 0)
mostrarCuentas(contadores)
```
